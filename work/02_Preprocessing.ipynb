```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/coooodek/requestLog/refs/heads/main/dataset/requestLog.csv')
df.head(2)
```

```python
# 요청일 컬럼 object -> datetime

df['요청일'] = pd.to_datetime(df.요청일)
```

```python
# 분기 컬럼 생성

df['분기'] = df['요청일'].dt.quarter
```

```python
# 연도 컬럼 생성

df['연도'] = df['요청일'].dt.year
```

```python
# 요청일 기준 중복 접수 건 삭제

df = df.drop_duplicates(subset = ['요청일'])
```

```python
# 제품그룹별 / 전체 접수 건 대비 비율 계산 (ETC로 그룹화 전)

df['접수비율'] = df['제품그룹'].map(df['제품그룹'].value_counts(normalize = True)*100)
```

```python
df['접수비율'].value_counts()
```

```python
df.loc[df['접수비율'] < 0.5, '제품그룹'] = 'ETC' # 0.5% 미만 접수 건 ETC로 분류
```

```python
df[df.제품그룹 == 'ETC']
```

```python
df.제품그룹.unique() # N,P,Q,T,U -> ETC
```

```python
# 제품그룹별 / 전체 접수 건 대비 비율 계산 (ETC로 그룹화 후)

df['접수비율'] = df['제품그룹'].map(df['제품그룹'].value_counts(normalize = True)*100)
```

```python
df[df['제품그룹'] == 'ETC'].head()
```

```python
df = df.drop(columns = '접수비율')
```
